## Visualization & Explanatory Analysis

Visualisasi data dan analisis eksplanatif merupakan aspek yang saling terkait dalam pemahaman dan komunikasi hasil analisis data.  Pertanyaan 1:Bagaimana persebaran customer berdasarkan customer city pada dataset customer?

```python
bystate_df = customers_df.groupby(by="customer_city").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
```

```python
bystate_df = bystate_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ukuran figur dan warna
plt.figure(figsize=(10, 6))
colors_ = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Menyesuaikan penamaan kolom agar sesuai dengan DataFrame yang digunakan
sns.barplot(
    x="customer_count",
    y="customer_city",
    data=bystate_df.sort_values(by="customer_count", ascending=False).head(10),
    palette=colors_
)
# Menambahkan judul dan pengaturan tambahan
plt.title("Top 10 Cities by Number of Customers", loc="center", fontsize=15)
plt.xlabel("Number of Customers")
plt.ylabel("City")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Pertanyaan 2: Produk apa saja yang paling banyak dan paling sedikit terjual?

```python
products_df.sample(10)
```

```python
products_df.head()
```

```python
sum_order_items_df = products_df.groupby("product_category_name").product_name_lenght.sum().sort_values(ascending=False).reset_index()
sum_order_items_df.head(10)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 8))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat subplot
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 8))

# Warna untuk plot
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Plot untuk produk terbaik (Best Performing Product)
sns.barplot(x="product_name_lenght", y="product_category_name",
            data=sum_order_items_df.sort_values(by="product_name_lenght", ascending=False).head(5),
            palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=15)

# Plot untuk produk terburuk (Worst Performing Product)
sns.barplot(x="product_name_lenght", y="product_category_name",
            data=sum_order_items_df.sort_values(by="product_name_lenght", ascending=True).head(5),
            palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

# Menambahkan judul keseluruhan
plt.suptitle("Best and Worst Performing Product by Number of Sales", fontsize=20)
plt.show()
```

Berdasarkan analisis dataset customer, dapat disimpulkan bahwa persebaran pelanggan (customer) bervariasi berdasarkan kota pelanggan (customer city).
